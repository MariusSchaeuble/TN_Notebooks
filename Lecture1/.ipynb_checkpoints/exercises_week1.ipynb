## Exercise 1

**1)** Create the indices necesary to describe the pure state of a quantum system consisting of a spin $1/2$ particle $S$ and a single bosonic mode $B$ with local dimension truncated to 3 (thus having only vacuum $\ket{0}$, singly excited $\ket{1}$ and doubly excited $\ket{2}$ states). The indices mus have tags `"Spin_1/2"` and `"Boson:n=3"`

```julia
using ITensors
using LinearAlgebra

```

```julia
i = Index(2, tags = "Spin_1/2")
j = Index(3, tags = "Boson:n=3")
```

**2)** Define the tensor with variable name `spinBoson` with the indices defined in the previous step.

```julia
spinBoson = ITensor(ComplexF64, i, j)
```

**3)** Initialize the state $\ket{\psi_0}  = \ket{\downarrow}_S \otimes \ket{1}_B$ 

```julia
spinState = ITensor(ComplexF64, i)
spinState[i=>1] = 1/sqrt(2)
spinState[i=>2] = -1/sqrt(2)
println(spinState)
bosonState = ITensor(ComplexF64, ComplexF64[0, 1, 0], j)
println(bosonState)

for a in range(1, 2), b in range(1, 3)
    spinBoson[i=>a, j=>b] = spinState[i=>a]*bosonState[j=>b]
end
println(spinBoson)
println(spinState*bosonState)
```

```julia
println(2*spinBoson)
```

## Exercise 2

**1)** Define function `evolve` that thakes in input:

1. An ITensor, pepresenting the state $\ket{\psi_0}$
2. A time $t$ (Float64)
3. A matrix representing the Hamiltonian (in the same basis as the one used for the state) and returns the evolved state $e^{-i t H} \ket{\psi_0}$.

__NOTE__: We ignore units; consider all the values as numbers.

```julia
function evolve(Ψ_0::ITensor, t::Float64, H::Matrix{ComplexF64})
    HT = ITensor(ComplexF64, H, i', j', i, j)
    evOp = exp(-1im*t*HT)
    return noprime(evOp*Ψ_0)
end
```

```julia
H_test = ComplexF64.(Matrix(Diagonal(ones(6))))
```

```julia
println(evolve(spinBoson, 1., H_test))
```

**2)** By using the function `evolve` monitor the evolution of the state $\ket{\psi_0}$ defined in the previous exercise by creating a an array of `ITensors` evolved under the action of the Hamitonian 
$$
H_1 = \frac \Delta 2 \sigma_x \otimes \omega a^\dagger a
$$
with $\Delta = 0.2$, $\omega = 0.4$ for a time $t_\text{max} = 20$ at steps $\delta t = 0.2$. 

```julia
Δ = 0.2
ω = 0.4

σ_x = ComplexF64[0 1; 1 0]
anzahlOp = ComplexF64[0 0 0;
0 1 0;
0 0 2]
H_1 = Δ*ω/2 *kron(σ_x, anzahlOp)
```

```julia
evolvedStates = ITensor[]
times = 0:0.2:20
for t in times
    push!(evolvedStates, evolve(spinBoson, t, H_1))
end
```

```julia
print(evolvedStates)
```

## Exercise 3

Using the array of evolved states (`ITensors`) created in the previous exercise, determine, store and plot:

1. the average values of $\sigma_x,\ \sigma_y$ and $\sigma_z$ of the spin,
2. the average occupation number ($N = a^\dagger a$) of the boson,

at each evolution step $t=0., 0.2,\ldots, 20.0$.

1. avg der spins

```julia
σ_x = ITensor(ComplexF64, ComplexF64[0 1; 1 0], i', i)
σ_y = ITensor(ComplexF64, ComplexF64[0 1im; -1im 0], i', i)
σ_z = ITensor(ComplexF64, ComplexF64[1 0; 0 -1], i', i)
σ_id = ITensor(ComplexF64, ComplexF64[1 0; 0 1], i', i)

idBoson = ITensor(ComplexF64, ComplexF64[1 0 0; 0 1 0; 0 0 1], j', j)

avg_x = zeros(ComplexF64, length(evolvedStates))
for l in range(1, length(evolvedStates))
    avg_x[l] = inner(evolvedStates[l]', σ_x * idBoson, evolvedStates[l])
end

avg_y = zeros(ComplexF64, length(evolvedStates))
for l in range(1, length(evolvedStates))
    avg_y[l] = inner(evolvedStates[l]', σ_y * idBoson, evolvedStates[l])
end

avg_z = zeros(ComplexF64, length(evolvedStates))
for l in range(1, length(evolvedStates))
    avg_z[l] = inner(evolvedStates[l]', σ_z * idBoson, evolvedStates[l])
end

avg_anzahl = zeros(ComplexF64, length(evolvedStates))
for l in range(1, length(evolvedStates))
    avg_anzahl[l] = inner(evolvedStates[l]', σ_id * ITensor(ComplexF64, anzahlOp, j', j), evolvedStates[l])
end

println(avg_x)
print(avg_anzahl)

using Plots
plot(times, real(avg_x))
plot!(times, real(avg_y))
plot!(times, real(avg_z))
plot!(times, real(avg_anzahl))
```

## Exercise 4

Repeat the operations of exercises 2 ad 3 using, this time, the Hamiltonian:

$$
H_{JC} = \frac{\epsilon}{2} \sigma_z + \omega a^\dagger a +  g(\sigma_+ a + \sigma_- a^\dagger)
$$

with $\epsilon = 0.2$, $\omega = 0.4$ and $g = 1.0$.

```julia
ϵ = 0.2
ω = 0.4
g = 10

σ_z_matrix = ComplexF64[1 0; -1 0]
id_boson_matrix = ComplexF64[1 0 0; 0 1 0; 0 0 1]
id_pauli_matrix = ComplexF64[1 0 ; 0 1]

σ_plus = ComplexF64[0 1; 0 0]
σ_minus = ComplexF64[0 0; 1 0]

ab_matrix = ComplexF64[0  1  0;
0  0  sqrt(2);
0  0  0]
auf_matrix = ab_matrix'


H_2 = ϵ/2*kron(σ_z_matrix, id_boson_matrix) + ω*kron(id_pauli_matrix, anzahlOp) + g*(kron(σ_plus, ab_matrix) + kron(σ_minus, auf_matrix))
```

```julia
evolvedStates2 = ITensor[]
times = 0:0.2:20
for t in times
    push!(evolvedStates2, evolve(spinBoson, t, H_2))
end
```

```julia
avg_x2 = zeros(ComplexF64, length(evolvedStates2))
for l in range(1, length(evolvedStates2))
    avg_x2[l] = inner(evolvedStates2[l]', σ_x * idBoson, evolvedStates2[l])
end

avg_y2 = zeros(ComplexF64, length(evolvedStates2))
for l in range(1, length(evolvedStates2))
    avg_y2[l] = inner(evolvedStates2[l]', σ_y * idBoson, evolvedStates2[l])
end

avg_z2 = zeros(ComplexF64, length(evolvedStates2))
for l in range(1, length(evolvedStates2))
    avg_z2[l] = inner(evolvedStates2[l]', σ_z * idBoson, evolvedStates2[l])
end

avg_anzahl2 = zeros(ComplexF64, length(evolvedStates2))
for l in range(1, length(evolvedStates2))
    avg_anzahl2[l] = inner(evolvedStates2[l]', σ_id * ITensor(ComplexF64, anzahlOp, j', j), evolvedStates2[l])
end

using Plots
plot(times, real(avg_x2))
plot!(times, real(avg_y2))
plot!(times, real(avg_z2))
plot!(times, real(avg_anzahl2))
```

## Exercise 5

Considering the data created in the previous exercise (Exercise 4), determine the reduced density matrix of the spin system at each evolution time.
By using the so obtained reduced states perform the measurements $\langle \sigma_{x,y,z} \rangle_{\rho_S} = \text{Tr}(\sigma_{x,y,z} \rho_S(t))$ and compare the results with those obtained in Ex. 4. 

Verwende evolvedstates2

```julia
rho_tot = []
for state in evolvedStates2
    push!(rho_tot, prime(state)*dag(state))
end
print(evolvedStates2[1])
print(rho_tot[1])

rho_spin = []
for element in rho_tot
    push!(rho_spin, element*delta(j', j))
end
print(rho_spin[1])

#mittelwerte
sigma_x_matrix = ComplexF64[0 1; 1 0]
sigma_y_matrix = ComplexF64[0 1im; -1im 0]
sigma_z_matrix = ComplexF64[1 0; 0 -1]


sigma_x_mean = []
for element in rho_spin
    tensor = ITensor(ComplexF64, sigma_x_matrix*matrix(element, i', i), i', i)
    push!(sigma_x_mean, scalar(tensor*delta(i', i)))
end
sigma_y_mean = []
for element in rho_spin
    tensor = ITensor(ComplexF64, sigma_y_matrix*matrix(element, i', i), i', i)
    push!(sigma_y_mean, scalar(tensor*delta(i', i)))
end
sigma_z_mean = []
for element in rho_spin
    tensor = ITensor(ComplexF64, sigma_z_matrix*matrix(element, i', i), i', i)
    push!(sigma_z_mean, scalar(tensor*delta(i', i)))
end

plot(times, real(sigma_x_mean))
plot!(times, real(sigma_y_mean))
plot!(times, real(sigma_z_mean))
```

## Exercise 6

Determine and plot the von Neumann entropy of the density matrix $\rho_S(t)$ at the times $t=0,0.2,\ldots,20.$.

```julia
#svd auf jedes element von rho_S 
sis = []
for denMatrix in rho_spin
    uu, s, vv = svd(denMatrix, i)# index i richtig?
    push!(sis, s)
end
#println(matrix(sis[10])[1][1])

# entropie berechnen:
Entr = []
for singL in sis
    entro = 0.
    for l in range(1, 2)
        entro += -1*matrix(singL)[l, l]*log(matrix(singL)[l, l])
    end
    push!(Entr, entro)
end
println(Entr[1])

plot(times, Entr)
```
