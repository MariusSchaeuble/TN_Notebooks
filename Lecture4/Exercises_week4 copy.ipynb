```julia
function leftCanonical(state)
    nn = length(state)
    uu, s, vv = svd(state[1], inds(state[1])[1])

    new2 = s*vv*state[2]
    state[1] = uu
    state[2] = new2
    
    for i in 2:nn-1
        uu, s, vv = svd(state[i], inds(state[i])[1:2])
        newIPlus = s*vv*state[i+1]
        state[i] = uu
        state[i+1] = newIPlus
    end
    return state
end
    
```

```julia
psi0
```

```julia
psiL = leftCanonical(psi0)
```

```julia
function rightCanonical(state)
    nn = length(state)
    uu, s, vv = svd(state[1], inds(state[nn])[2])

    new2M = s*uu*state[nn-1]
    state[nn] = vv
    state[nn-1] = new2M
    
    for i in nn-1:-1:2
        uu, s, vv = svd(state[i], inds(state[i])[2:3])
        newIPlus = s*uu*state[i-1]
        state[i] = vv
        state[i-1] = newIPlus
    end
    return state
end
```

```julia
psiR = rightCanonical(psi0)
```
