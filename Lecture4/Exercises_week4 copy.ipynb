```julia
function leftCanonical(state)
    nn = length(state)
    uu, s, vv = svd(state[1], inds(state[1])[1])

    new2 = s*vv*state[2]
    state[1] = uu
    state[2] = new2
    
    for i in 2:nn-1
        uu, s, vv = svd(state[i], inds(state[i])[1:2])
        newIPlus = s*vv*state[i+1]
        state[i] = uu
        state[i+1] = newIPlus
    end
    return state
end
    
```

```julia
function rightCanonical(state)
    nn = length(state)
    uu, s, vv = svd(state[nn], commonind(state[nn], state[nn-1]))

    new2M = s*uu*state[nn-1]
    state[nn] = vv
    state[nn-1] = new2M
    
    for i in nn-1:-1:2
        uu, s, vv = svd(state[i], commonind(state[i], state[i-1]))
        newIPlus = s*uu*state[i-1]
        state[i] = vv
        state[i-1] = newIPlus
    end
    return state
end
```

**1)** Write a function that, given an input MPS compute the expectation value of $S_z$ on every sites;

**2)** Create a random MPS with $N = 10$ spins and a bond dimension $D=4$ and measure $S_z$ with the function that you wrote in **1)**; 

**3)** Within `ITensor`, the expectation values of a local observable can be computed using the `expect` function that takes as an input a MPS and an operator. Compare the results you obbtained in **2)** with the function `expect(psi,"Sz")` that returns the expectation of $S_z$ on each site.

```julia
function expectSZ(state)
    indis = siteinds(state)
    nn = length(state)
    opsum = OpSum()
    for i in 1:nn
        opsum += "Sz", i
    end
    operatorSZ = MPO(opsum, indis)

    return inner(state', operatorSZ, state)
end
```

**3)** The function `correlation_matrix(psi,"Sz","Sz")` from `ITensors` returns the correlation matrix $\bra{\psi}S^z_i S^z_j \ket{\psi}$. Compare your results with `correlation_matrix`.

```julia
function correl1I(state, i)
    indis = siteinds(state)
    nn = length(state)
    opsum = OpSum()
    opsum += "Sz", 1, "Sz", i
    operatorSZ = MPO(opsum, indis)

    return inner(state', operatorSZ, state)
end
```

## Exercise 4: Measurements as a Convergence Check

Last week, in [MPO_ITensor.ipynb](../Lecture3/MPO_ITensor.ipynb), you used the 3rd order expansion of the evolution operator $exp(-i H \delta t)$ to compute the time-evolution of a MPS. This is an approximation that depends on the magnitude of the Hamiltonian matrix and the time-step $\delta t$. 
We know that for a closed system, the energy $\langle H \rangle$ is conserved. Thus, this expectation value can be used as a convergence check. If it doesn't stay constant, one should reduce $\delta t$ and/or add orders to the expansion.

**1)** Compute the expectation value of the energy at each time step for an evolution using a 2nd order expansion;

**2)** Compute the expectation value of the energy at each time step for an evolution using a 3rd order expansion and play with the value of $\delta t$.

```julia
nn = 20
jj=1.0
system = siteinds("S=1/2",nn)
#Prepare the initial state as a product state with "Up" on the first site and "Dn" on the rest of the sites
psi0= MPS(ComplexF64,system,[(j==1) ? "Up" : "Dn" for j in 1:nn]);
psi0copy = deepcopy(psi0)
```

```julia
# define a hamiltonian
h = -0.5
opsum = OpSum()
for bond in 1:nn-1
    #note that the sintax allows for constants to appear before the operator string
    opsum += -1. * jj, "S-", bond, "S+", bond+1
    opsum += -1. * jj, "S+", bond, "S-", bond+1
end

ham = MPO(opsum, system)
```

```julia
deltat = 0.5
nsteps = Int(round(10/deltat))
hdeltat = deltat * ham
#build an MPO with the operator "Id" on each site
theOne = MPO(system,"Id")
factor(k) = (-1im)^k / factorial(k)
#As not to have all the indices of hdeltat contracted with each other, we prime the second one and then we contract the first index of the second one with the first index of the first one, and then we lower the second prime level to first prime level.
h2 = mapprime(prime(hdeltat)*hdeltat,2 =>1)
h3 = mapprime(prime(hdeltat) * h2,2 =>1)
terms = [theOne, factor(1)*hdeltat, factor(2)*h2, factor(3)*h3];
appo = terms[1];
for t in terms[2:end]
    appo = t + appo
end
h3thOrder = appo;

terms = [theOne, factor(1)*hdeltat, factor(2)*h2];
appo = terms[1];
for t in terms[2:end]
    appo = t + appo
end
h2thOrder = appo;
#h4thOrder = +(terms...; cutoff=1e-18);
```

```julia
evolved3 = Vector{MPS}()
push!(evolved3, psi0)
for j=2:nsteps
    push!(evolved3, noprime(h3thOrder * evolved3[end]))
end
```

```julia
function expectH(state, ham)
    return inner(state', ham, state)
end
```

```julia
for state in evolved3
    println(expectH(state, h3thOrder))
end
```

```julia
evolved2 = Vector{MPS}()
push!(evolved2, psi0)
for j=2:nsteps
    push!(evolved2, noprime(h2thOrder * evolved2[end]))
end
```

```julia
for state in evolved2
    println(expectH(state, h2thOrder))
end
```
