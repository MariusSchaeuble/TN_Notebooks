## Exercise 4: Measurements as a Convergence Check

Last week, in [MPO_ITensor.ipynb](../Lecture3/MPO_ITensor.ipynb), you used the 3rd order expansion of the evolution operator $exp(-i H \delta t)$ to compute the time-evolution of a MPS. This is an approximation that depends on the magnitude of the Hamiltonian matrix and the time-step $\delta t$. 
We know that for a closed system, the energy $\langle H \rangle$ is conserved. Thus, this expectation value can be used as a convergence check. If it doesn't stay constant, one should reduce $\delta t$ and/or add orders to the expansion.

**1)** Compute the expectation value of the energy at each time step for an evolution using a 2nd order expansion;

**2)** Compute the expectation value of the energy at each time step for an evolution using a 3rd order expansion and play with the value of $\delta t$.
